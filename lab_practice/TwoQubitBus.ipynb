```python
import qiskit_metal as metal
from qiskit_metal import designs, draw
from qiskit_metal import MetalGUI, Dict
```

```python
design = designs.DesignPlanar()
gui = MetalGUI(design)
```

```python
from qiskit_metal.qlibrary.qubits.transmon_pocket_6 import TransmonPocket6

design.overwrite_enabled = True
design.chips.main
design.chips.main.size.size_x = '11mm'
design.chips.main.size.size_y = '9mm'
```

```python
Q1 = TransmonPocket6(design,'Q1', options = dict(
        pos_x='-2mm', 
        pos_y='0mm', 
        gds_cell_name ='FakeJunction_01',
        hfss_inductance ='14nH',
        connection_pads = dict(
            bus_q1_q2 = dict(loc_W=1, loc_H=-1, pad_width = '80um', pad_gap = '15um'),
            readout = dict(loc_W=0, loc_H=1, pad_width = '80um', pad_gap = '50um'))))

Q2 = TransmonPocket6(design,'Q2', options = dict(
        pos_x='2mm', 
        pos_y='0mm', 
        gds_cell_name ='FakeJunction_01',
        hfss_inductance ='14nH',
        connection_pads = dict(
            bus_q1_q2 = dict(loc_W=-1, loc_H=-1, pad_width = '80um', pad_gap = '15um'),
            readout = dict(loc_W=0, loc_H=1, pad_width = '80um', pad_gap = '50um'))))

gui.rebuild()
gui.autoscale()
```

```python
from qiskit_metal.analyses.em.cpw_calculations import guided_wavelength

def find_resonator_length(frequency, line_width, line_gap, N): 
    #frequency in GHz
    #line_width/line_gap in um
    #N -> 2 for lambda/2, 4 for lambda/4
    
    [lambdaG, etfSqrt, q] = guided_wavelength(frequency*10**9, line_width*10**-6,
                                              line_gap*10**-6, 750*10**-6, 200*10**-9)
    return str(lambdaG/N*10**3)+" mm"
```

```python
from qiskit_metal.qlibrary.tlines.meandered import RouteMeander

bus_q1_q2 = RouteMeander(design,'Bus_Q1_Q2', options = dict(hfss_wire_bonds = True, 
                                            pin_inputs=Dict(
                                                start_pin=Dict(
                                                    component='Q1',
                                                    pin='bus_q1_q2'),
                                                end_pin=Dict(
                                                    component='Q2',
                                                    pin='bus_q1_q2')
                                            ),
                                            lead=Dict(
                                                start_straight='125um',
                                                end_straight = '125um'
                                            ),
                                            meander=Dict(
                                                asymmetry = '50um'),
                                            fillet = "99um",
                                            total_length = '7.4mm'))

gui.rebuild()
gui.autoscale()
```

```python
from qiskit_metal.qlibrary.terminations.launchpad_wb import LaunchpadWirebond

launch_nw = LaunchpadWirebond(design, 'Launch_NW',options = dict(pos_x = '-5mm', pos_y='2mm', orientation=0))
launch_ne = LaunchpadWirebond(design, 'Launch_NE',options = dict(pos_x = '5mm', pos_y='2mm', orientation=180))

gui.rebuild()
gui.autoscale()
```

```python
from qiskit_metal.qlibrary.couplers.coupled_line_tee import CoupledLineTee

q1_read_T = CoupledLineTee(design,'Q1_Read_T', options=dict(pos_x = '-2mm', pos_y = '2mm', 
                                                        orientation = '0', 
                                                        coupling_length = '200um',
                                                        open_termination = False))

q2_read_T = CoupledLineTee(design,'Q2_Read_T', options=dict(pos_x = '2mm', pos_y = '2mm', 
                                                        orientation = '0', 
                                                        coupling_length = '200um',
                                                        open_termination = False))

gui.rebuild()
gui.autoscale()
```
